```python
import matplotlib.pyplot as plt
```

```python
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
```

```python
data.dropna(subset=['date_added'], inplace=True)
```

```python
print("Generating Year-wise Series and Movies Comparison plot...")
```

```python
data['year_added'] = data['date_added'].dt.year
```

```python
content_by_year_type = data.groupby(['year_added', 'type']).size().unstack(fill_value=0)
plt.figure(figsize=(14, 8))
content_by_year_type.plot(kind='bar', stacked=False, colormap='viridis', ax=plt.gca())
plt.title('Number of Movies vs. TV Shows Added to Netflix by Year')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Content Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print("Generating 2015-2020 Comedy vs. Thriller Movies Pie Chart")
```

```python
movies_2015_2020 = data[
    (data['type'] == 'Movie') &
    (data['year_added'] >= 2015) &
    (data['year_added'] <= 2020)
].copy()
```

```python
exploded_genres = movies_2015_2020.assign(listed_in=movies_2015_2020['listed_in'].str.split(', ')).explode('listed_in')
```

```python
comedy_thriller_movies = exploded_genres[
    exploded_genres['listed_in'].isin(['Comedies', 'Comedy', 'Thrillers', 'Thriller'])
].copy()
```

```python
comedy_thriller_movies['standardized_genre'] = comedy_thriller_movies['listed_in'].replace({
    'Comedies': 'Comedy',
    'Thrillers': 'Thriller'
})
```

```python
genre_counts = comedy_thriller_movies['standardized_genre'].value_counts()
```

```python
plt.figure(figsize=(9, 9))
plt.pie(
            genre_counts,
            labels=genre_counts.index,
            autopct='%1.1f%%', # Show percentage on slices
            startangle=90,
            colors=['gold', 'lightcoral'] # Custom colors for clarity
        )
plt.title('Proportion of Comedy vs. Thriller Movies Added to Netflix (2015-2020)')
plt.axis('equal')  # Equal aspect ratio 
plt.tight_layout()
plt.show()
```
